```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from skimage.metrics import structural_similarity
```

```
# Read the image
image1 = cv2.imread("image path1")
image2 = cv2.imread("image path2")
width,height,channel=image1.shape 

gray1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)# Conversion to grayscale
height_1,width_1,=gray1.shape
gray2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)# Conversion to grayscale
height_2,width_2=gray2.shape
#gray3 = cv2.cvtColor(r_image, cv2.COLOR_BGR2GRAY)
print(height_1,width_1)
print(height_2,width_2)
```

```
if width_1 < width_2 :  # resizing to same size for comparison
    final_width=width_1
else:

    final_width=width_2

if height_1<height_2:
    final_height=height_1
else :
    final_height=height_2
 
print (final_width,final_height)
```

```
#Rescaling
scaled_image1 = cv2.resize(gray1, (final_width,final_height))
scaled_image2 = cv2.resize(gray2, (final_width,final_height))

cv2_imshow(scaled_image1)
cv2_imshow(scaled_image2)
```

```
# Score check
(score, diff) = structural_similarity(scaled_image1, scaled_image2, full=True)
diff = (diff * 255).astype("uint8")
print("SSIM: {}".format(score))
```
